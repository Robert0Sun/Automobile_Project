```python

#We will be importing pandas and numpy for our numerical analysis
#We will be importing pyplot and seaborn in order to visualize our data

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load data as a pandas dataframe

df = pd.read_csv('/Users/rsuncin.fellow/Documents/GitHub/automobile/imports-85.csv')
```

```python
#We see the dataframe does not have their columns name so, lets change that
#also rename the dataframe
automobiles = df

automobiles
```

```python
#here i am naming he unnamed columns to the appropaite names

automobiles.columns = ["Symboling", "Normalized_Loses", "Make", "Fuel_type", "Aspiration", "Num_of_Doors", "Body_style", 
                       "Drive_wheels", "Engine_location", "Wheel_base", "Length", "Width", "Height", "Curb_weight", "Engine_type", 
                       "Num_of_cylinders", "Engine_size", "Fuel_system", "Bore", "Stroke", "Compression_ratio", "Horsepower", "Peak_RPM", 
                       "City_MPG", "Highway_MPG", "Price"]

automobiles
```

```python
#Rename the columns through a dictionar method

automobiles.rename(columns = {"Symboling" : "Symboling", "Normalized_Loses" : "Nomralized_loses", "Make" : "Make", "Fuel_type" : "Fuel_type", "Aspiration" : "Aspiration", "Num_of_Doors" : "Num_Doors", "Body_style" : "Body_style", 
                       "Drive_wheels" : "Drive_wheels", "Engine_location" : "Engine_loc", "Wheel_base" : "Wheel_base", "Length" : "Length", "Width" : "Width", "Height" : "Height", "Curb_weight" : "Curb_weight", "Engine_type" : "Engine_type", 
                       "Num_of_cylinders" : "Num_cylinders", "Engine_size" : "Engine_size", "Fuel_system" : "Fuel_system", "Bore" : "Bore", "Stroke" : "Stroke", "Compression_ratio" : "Compression_ratio", "Horsepower" : "Horsepower", "Peak_RPM" : "Peak_RPM", 
                       "City_MPG" : "City_MPG", "Highway_MPG" : "Highway_MPG", "Price" : "Price"}, inplace = True)

automobiles
```

```python
#checking to see if there are anymissing data
#data does have missing values but for some reason this list isnt reflecting that

automobiles.isnull().sum()
```

```python
#missing values were shown as '?' thats why it was not popping up as a missing value
#gotta replace the '?' with pythons defaul marker for missing values

automobiles.replace("?", np.nan, inplace = True)
automobiles
```

```python
#now i check to see if i can see how many values are actually missing
#now we can see that the catagory 'Normalized_loses' has most values missing

automobiles.isnull().sum()


```

```python
#find the average of the columns and then replace the missing value with the Average value


avg_norm_loss = automobiles['Nomralized_loses'].astype("float").mean(axis=0)
print('Average of Nomralized-loses:', avg_norm_loss)

avg_bore = automobiles['Bore'].astype("float").mean(axis=0)
print('Average of Bore:', avg_bore)  

avg_stroke = automobiles['Stroke'].astype("float").mean(axis=0)
print('Average of Stroke:', avg_stroke)  

avg_horsepower = automobiles['Horsepower'].astype("float").mean(axis=0)
print('Average of Horsepower:', avg_horsepower)  

avg_peak_rpm = automobiles['Peak_RPM'].astype("float").mean(axis=0)
print('Average of Peak_RPM:', avg_peak_rpm)  

avg_price = automobiles['Price'].astype("float").mean(axis=0)
print('Average of Price:', avg_price)  


automobiles['Nomralized_loses'].replace(np.nan, avg_norm_loss, inplace=True) 
automobiles['Bore'].replace(np.nan, avg_bore, inplace=True) 
automobiles['Stroke'].replace(np.nan, avg_stroke, inplace=True)
automobiles['Horsepower'].replace(np.nan, avg_horsepower, inplace=True) 
automobiles['Peak_RPM'].replace(np.nan, avg_peak_rpm, inplace=True) 
automobiles['Price'].replace(np.nan, avg_price, inplace=True)  
```

```python
#I am going to split that data into one section being catagorical column and the other the number column
#and then from there start graphing each data to see the outliers

category_columns = automobiles.select_dtypes(include = ['object']).columns
number_columns = automobiles.select_dtypes(include = np.number).columns.tolist()

print("Categorical Variables:")
print(category_columns)
print("Numerical Variables:")
print(number_columns)
```

```python
sns.pairplot(automobiles[number_columns])
```

```python
for col in number_columns:
    print(col)
    print('Skew :', round(automobiles[col].skew(), 2))
    plt.figure(figsize = (15,4))
    plt.subplot(1,2,1)
    automobiles[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1,2,2)
    sns.boxplot(x = automobiles[col])
    plt.show()
```
